**Importing necessary libraries**

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

**Numerical - Bivariate**

```
#Polynomial plots

x1 = [1, 3, 5, 7, 9]
y1 = [2, 4, 6, 8, 10]
x2 = [1, 2, 3, 4, 5]
y2 = [2, 8, 18, 32, 50]
plt.plot(x1, y1, label="The first plot")
plt.plot(x2, y2, label="The second plot")
plt.legend()
plt.xlabel("The X label")
plt.ylabel("The Y label")
plt.show()
```

**Categorical - Numerical(Bivariate)**

```
#According to the seaborn documentation, a bar plot represents an estimation of central tendency for a numerical variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars.
#Bar plots include 0 in the quantitative axis range, and they are a good choice when 0 is a meaningful value for the quantitative variable, and you want to make comparisons against it.
#The distribution of some categorical variable around a mean value

plt.bar(x1, y1, label="This is the bar plot for first function", color='green')
plt.bar(x2, y2, label="This is the bar plot for second function", color="orange")
plt.title("This is bar plot")
plt.xlabel("This is x label")
plt.ylabel("This is y label")
plt.legend()
plt.show()
```

**Numerical/Categorical-Univariate**

```
#In statistics, the Freedman–Diaconis rule can be used to select the width of the bins to be used in a histogram.[1] It is named after David A. Freedman and Persi Diaconis.
#Optimal Bin width is decided as = (2 * IQR(x))/(n^1/3) - Freedman-Diaconsis rule

ages = [22, 23, 45, 56, 22, 45, 56, 67, 89, 89, 90, 44, 28, 90, 65, 13, 13, 65, 56]
plt.hist(ages, bins=15,  rwidth=2)
plt.show()
```

**Numerical Bivariate**

```
#Scatter plots

plt.style.use("fivethirtyeight")

plt.scatter(x1, y1, label="First dataset", color="indigo", s=25)
plt.scatter(x2, y2, label="Second dataset", color="blue", s=30)

plt.xlabel("First Dataset")
plt.ylabel("Second Dataset")
plt.title("Scatter Plot")
plt.legend(loc="upper right")

plt.tight_layout()

plt.show()
```

**Numerical multivariate**

```
#Stack plots. Similar to pie plots in the sense that even these are a measure of proportions, but unlike a pie plot which shows it in a single timeframe, stack plots are used to 
#Represent plots over a timeframe
#Best used for presenting summary statistics by statistics
#Depicting the sleeping pattern of my bros over a period of time
#Keeping stack of the amount of hours each one is dedicating to building the project

plt.style.use("fivethirtyeight")

minutes = [1, 2, 3, 4, 5, 6, 7, 8, 9]
sagar_1 = [1, 1, 3, 3, 3, 4, 6, 8, 9]
saumy_1 = [1, 1, 2, 3, 2, 5, 5, 6, 7]
rachi_1 = [1, 1, 1, 2, 2, 2, 4, 5, 7]

labels = ['Sagar', 'Saumya', 'Rachit']
plt.stackplot(minutes, sagar_1, saumy_1, rachi_1, colors=['r', 'g', 'b'], labels=labels)
plt.title("My basic stack plot")
plt.legend(loc="upper left")

plt.tight_layout()

plt.show()
```

**Plotting legends**

```
#Plotting legends externally
y2 = [10, 11, 12, 13, 16]

plt.stackplot(x1, y1, y2, colors=['r', 'g', 'b'])
plt.plot([], [], label='This is how we can apply the label', color='b')

plt.title("This is stackplot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend(loc="upper left")
plt.show()
```

```
#SLices in a single timeframe. That's a piechart. Ishan is probably going to scream upon discovering his low share of pizza slice. 
slices=[90, 40, 30, 20]
activities = ["Sagar", "Ishan", "Unknown1", "Unknown2"]
plt.pie(slices, labels=activities, explode=[.1, .1, .1, .1], colors=['r', 'g', 'b', 'y'], shadow=True, autopct="%1.1f%%", startangle=25)
plt.show()
```

```
#Custom function plots

def f(t):
  return np.exp(-t) + np.cos(2 * np.pi * t)

t1 = np.arange(0.0, 5.0, 0.01)
t2 = np.arange(0.0, 5.0, 0.02)
t3 = np.arange(0.9, 5.0, 0.07)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k', t3, f(t3), '--')

plt.subplot(212)
plt.plot(t1, np.cos(2 * np.pi * t1), 'r+')

plt.show()
```
